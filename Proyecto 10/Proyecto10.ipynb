### Matriz de inercia

```python
import seaborn as sns
import plotly.express as px

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
X = S.filter(['abigeato1', 'amenazas1', 'delitos_sexuales1', 'extorsion1',
       'homicidios1', 'homicidios_en_accidente_de_transito1',
       'hurto_a_comercio1', 'hurto_a_entidades_financieras1',
       'hurto_a_personas1', 'hurto_a_residencias1', 'hurto_automotores1',
       'hurto_motocicletas1', 'lesiones_en_accidente_de_transito1',
       'lesiones_personales1', 'pirateria_terrestre1', 'secuestro1',
       'terrorismo1', 'violencia_intrafamiliar1']); X
```

```python
X.corr()
```

```python
Municipios = S.filter(['muni'])
Indicador_1 = X.index ; Mun = Municipios.to_numpy()
dic_1 = {}
for j in range(X.shape[0]):
  dic_1[Indicador_1[j]] = Mun[j]
X.rename(index = dic_1)
```

```python
X_std = StandardScaler().fit_transform(X); X_std
```

#### A mano

```python
cov = X_std.T.dot(X_std)/X_std.shape[0]
```

```python
eigvals, eigvecs = np.linalg.eig(cov)
```

```python
varianza_explicada = {i:i/sum(eigvals) for i in (eigvals)} ;varianza_explicada
```

```python
inercia = pd.DataFrame({'eigvals':eigvals,
                       'inercia_acum': np.cumsum(eigvals),
                       'var_expl': varianza_explicada.values(),
                       'v_expl_acum':np.cumsum([i for i in varianza_explicada.values()])})
```

```python
inercia.T.round(4)
```

## Con scikit-learn 

```python
pca = PCA()
X_pca = pca.fit_transform(X_std)
```

```python
eigvals = pca.explained_variance_ ; eigvals
```

```python
inercia_acum = eigvals.cumsum(); inercia_acum
```

```python
var_expl = pca.explained_variance_ratio_ ; var_expl
```

```python
v_expl_acum = var_expl.cumsum(); v_expl_acum
```

```python
inercia_matrix = pd.DataFrame({'eigvals':eigvals,
                               'inercia_acum':inercia_acum,
                               'var_expl': var_expl,
                               'v_expl_acum':v_expl_acum})
```

```python
inercia_matrix.T.round(4)
```

#### A mano

```python
X2_std = StandardScaler().fit_transform(X2); 
```

```python
cov = X2_std.T.dot(X2_std)/X2_std.shape[0]
```

```python
eigvals, eigvecs = np.linalg.eig(cov)
```

```python
varianza_explicada = {i:i/sum(eigvals) for i in (eigvals)} ;varianza_explicada
```

```python
inercia = pd.DataFrame({'eigvals':eigvals,
                       'inercia_acum': np.cumsum(eigvals),
                       'var_expl': varianza_explicada.values(),
                       'v_expl_acum':np.cumsum([i for i in varianza_explicada.values()])})
```

```python
inercia.T.round(4)
```

#### Con scikit-learn

```python
pca2 = PCA()
X_pca2 = pca.fit_transform(X2_std)
```

```python
eigvals = pca.explained_variance_ ; eigvals
```

```python
inercia_acum = eigvals.cumsum(); inercia_acum 
```

```python
var_expl = pca.explained_variance_ratio_ ; var_expl 
```

```python
v_expl_acum = var_expl.cumsum(); v_expl_acum 
```

```python
inercia_matrix = pd.DataFrame({'eigvals':eigvals, 
                               'inercia_acum':inercia_acum, 
                               'var_expl': var_expl, 
                                'v_expl_acum':v_expl_acum}) 
```

```python
inercia_matrix.T.round(4)
```
